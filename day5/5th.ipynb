```python
# Generate 250 random samples from a normal distribution with mean 170 and std 10
x_random = np.random.normal(170, 10, 250)
plt.hist(x_random)
plt.title('Histogram of Random Normal Distribution')
plt.show()
```

```python
#generate data for a box plot
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True, labels=['A', 'B', 'C'])
plt.title('Box Plot of Random Normal Distributions')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

```python
values  = [30, 40, 30]
labels = ['caregory A', 'category B', 'category C']
explode = (0.1, 0, 0)  # only "explode" the 1st slice
myColors = ['gold', 'yellowgreen', 'lightcoral']
plt.pie(values, explode=explode, labels=labels, colors=myColors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart of Categories')
plt.show()
```

```python
#sample data for energy consumption over 6 months in MWh and plot a line plot
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [150, 240, 350, 290, 350, 400]
plt.plot(months, energy_consumption, marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (MWh)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#energy consumption for renewable energy sources and plot a bar chart
renewable_sources = ['Solar', 'Wind', 'Hydro', 'Geothermal']
energy_consumption = [120, 200, 150, 80]
plt.bar(renewable_sources, energy_consumption, color=['gold', 'lightblue', 'green', 'brown'])
plt.title('Energy Consumption by Renewable Sources')
plt.xlabel('Renewable Energy Sources')
plt.ylabel('Energy Consumption (MWh)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#draw a pie chart and apply explode to it for renewable energy sources and also mention start angle
values = [120, 200, 150, 80]
labels = ['Solar', 'Wind', 'Hydro', 'Geothermal']
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice
myColors = ['gold', 'lightblue', 'green', 'brown']
plt.pie(values, explode=explode, labels=labels, colors=myColors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart of Renewable Energy Sources')
plt.show()
```

```python
import seaborn as sns
sns.set_theme()
fmri = sns.load_dataset("fmri")
sns.relplot( kind="line", x="timepoint", y="signal" , col = "region", hue="event", style="event", data=fmri)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('Scatter plot: total bill vs tip')
plt.show()
sns.lineplot(x="total_bill", y="tip", hue="day", style="day", data=tips)
plt.title('Line plot: total bill vs tip')
plt.show()
sns.barplot(x="day", y="total_bill", hue="sex",  data=tips)
plt.title('Bar plot: total bill by day')
plt.show()
```
