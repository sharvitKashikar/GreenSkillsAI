```python
class EnergySystems:
    def __init__(self, building_name, emission_fac, energy_cons):
        self.building_name = building_name
        self.emission_fac = emission_fac
        self.energy_cons = energy_cons

    def calculate_cf(self):
        return self.emission_fac * self.energy_cons

# Create an instance
A = EnergySystems("Building A", 0.28, 105)

# Print the details and carbon footprint
print(A.building_name, A.emission_fac, A.energy_cons)
print(f"Carbon Footprint: {A.calculate_cf():.2f} kg CO2")
```

```python
class Waste:
    def __init__(self, weight_kg, category):
        self.weight_kg = weight_kg
        self.category = category
    
    def get_weight(self):
        return self.weight_kg
    
    def get_category(self):
        return self.category

class PlasticWaste(Waste):
    def __init__(self, weight_kg, category, type_of_plastic):
        super().__init__(weight_kg, category)
        self.type_of_plastic = type_of_plastic
    
    def get_plastic_type(self):
        return self.type_of_plastic
    
    def is_recyclable(self):
        recyclable_types = ['PET', 'HDPE', 'PVC', 'LDPE', 'PP', 'PS']
        return self.type_of_plastic in recyclable_types


if __name__ == "__main__":
   
    general_waste = Waste(5.0, "biodegradable")
    print(f"General Waste - Weight: {general_waste.get_weight()}kg, Category: {general_waste.get_category()}")
    
  
    plastic_bottle = PlasticWaste(0.5, "recyclable", "PET")
    print(f"\nPlastic Waste - Weight: {plastic_bottle.get_weight()}kg")
    print(f"Category: {plastic_bottle.get_category()}")
    print(f"Type of Plastic: {plastic_bottle.get_plastic_type()}")
    print(f"Is Recyclable: {plastic_bottle.is_recyclable()}")
    
   
    plastic_bag = PlasticWaste(0.1, "non-recyclable", "OTHER")
    print(f"\nPlastic Waste - Weight: {plastic_bag.get_weight()}kg")
    print(f"Category: {plastic_bag.get_category()}")
    print(f"Type of Plastic: {plastic_bag.get_plastic_type()}")
    print(f"Is Recyclable: {plastic_bag.is_recyclable()}")
```
